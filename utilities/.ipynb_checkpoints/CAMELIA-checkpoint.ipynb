```python
all_perfs = []
eps=1e-6
for eval_mode in ['interpolation','extrapolation']:
    for fold in range(1,5):
        P_train, Y_train, P_test, Y_test = get_species_data(metadata,Y,fold=fold,eval_mode=eval_mode)
        
        ### Raw probability: threshold 0.5
        train_perfs = eval_species_perfs(Ys_hat=P_train,Ys_true=Y_train, th_list=0.5)
        train_perfs['fold']=fold
        train_perfs['eval_mode']=eval_mode
        train_perfs['subset']='train'
        train_perfs['postprocessing']='raw'
        all_perfs.append(train_perfs)
        del train_perfs
        
        test_perfs = eval_species_perfs(Ys_hat=P_test,Ys_true=Y_test, th_list=0.5)
        test_perfs['fold']=fold
        test_perfs['eval_mode']=eval_mode
        test_perfs['subset']='test'
        test_perfs['postprocessing']='raw'
        all_perfs.append(test_perfs)
        del test_perfs
        
        ### With optimized thresholds
        th_list = optimize_thres(Y_train, P_train)
        train_perfs = eval_species_perfs(Ys_hat=P_train,Ys_true=Y_train, th_list=th_list)
        train_perfs['fold']=fold
        train_perfs['eval_mode']=eval_mode
        train_perfs['subset']='train'
        train_perfs['postprocessing']='opth'
        all_perfs.append(train_perfs)
        del train_perfs

        test_perfs = eval_species_perfs(Ys_hat=P_test,Ys_true=Y_test, th_list=th_list)
        test_perfs['fold']=fold
        test_perfs['eval_mode']=eval_mode
        test_perfs['subset']='test'
        test_perfs['postprocessing']='opth'
        all_perfs.append(test_perfs) 
        del test_perfs
        
        ### With optimized normalized thresholds
        th_list = optimize_thres(Y_train, P_train)
        hat_P_train = (P_train - np.array(th_list))/(1 - np.array(th_list))
        hat_P_train = hat_P_train.clip(lower=0,upper=1)
        train_perfs = eval_species_perfs(Ys_hat=hat_P_train,Ys_true=Y_train, th_list=th_list)
        train_perfs['fold']=fold
        train_perfs['eval_mode']=eval_mode
        train_perfs['subset']='train'
        train_perfs['postprocessing']='opth_norm'
        all_perfs.append(train_perfs)
        del train_perfs
        
        
        hat_P_test = (P_test - np.array(th_list))/(1 - np.array(th_list))
        hat_P_test = hat_P_test.clip(lower=0,upper=1)
        test_perfs = eval_species_perfs(Ys_hat=hat_P_test,Ys_true=Y_test, th_list=th_list)
        test_perfs['fold']=fold
        test_perfs['eval_mode']=eval_mode
        test_perfs['subset']='test'
        test_perfs['postprocessing']='opth_norm'
        all_perfs.append(test_perfs) 
        del test_perfs

        
        ### Using PRR rule
        filter_prr = (P_train.rank(axis=1,method='min',ascending=False)<=Y_train.sum(axis=1).values.reshape(-1,1))*1+eps
        Y_hat_prr = P_train*filter_prr   

        ##### MASK
        train_perfs = eval_species_perfs(Ys_hat=filter_prr,Ys_true=Y_train, th_list=eps)
        train_perfs['fold']=fold
        train_perfs['eval_mode']=eval_mode
        train_perfs['subset']='train'
        train_perfs['postprocessing']='prr_mask'
        all_perfs.append(train_perfs)
        del train_perfs

        ##### CLIPPED PROBA
        train_perfs = eval_species_perfs(Ys_hat=Y_hat_prr,Ys_true=Y_train, th_list=eps)
        train_perfs['fold']=fold
        train_perfs['eval_mode']=eval_mode
        train_perfs['subset']='train'
        train_perfs['postprocessing']='prr'
        all_perfs.append(train_perfs) 
        del train_perfs

        filter_prr = (P_test.rank(axis=1,method='min',ascending=False)<=Y_test.sum(axis=1).values.reshape(-1,1))*1+eps
        Y_hat_prr = P_test*filter_prr   

        ##### MASK
        test_perfs = eval_species_perfs(Ys_hat=filter_prr,Ys_true=Y_test, th_list=eps)
        test_perfs['fold']=fold
        test_perfs['eval_mode']=eval_mode
        test_perfs['subset']='test'
        test_perfs['postprocessing']='prr_mask'
        all_perfs.append(test_perfs)
        del test_perfs

        ##### CLIPPED PROBA
        test_perfs = eval_species_perfs(Ys_hat=Y_hat_prr,Ys_true=Y_test, th_list=eps)
        test_perfs['fold']=fold
        test_perfs['eval_mode']=eval_mode
        test_perfs['subset']='test'
        test_perfs['postprocessing']='prr'
        all_perfs.append(test_perfs)  
        del test_perfs
        
        
        ### Using PRR rule
        th_list = optimize_thres(Y_train, P_train)
        hat_P_train = (P_train - np.array(th_list))/(1 - np.array(th_list))
        hat_P_train = hat_P_train.clip(lower=0,upper=1)
        
        filter_prr = (hat_P_train.rank(axis=1,method='min',ascending=False)<=Y_train.sum(axis=1).values.reshape(-1,1))*1+eps
        Y_hat_prr = hat_P_train*filter_prr   

        ##### MASK
        train_perfs = eval_species_perfs(Ys_hat=filter_prr,Ys_true=Y_train, th_list=eps)
        train_perfs['fold']=fold
        train_perfs['eval_mode']=eval_mode
        train_perfs['subset']='train'
        train_perfs['postprocessing']='prr_mask'
        all_perfs.append(train_perfs)
        del train_perfs

        ##### CLIPPED PROBA
        train_perfs = eval_species_perfs(Ys_hat=Y_hat_prr,Ys_true=Y_train, th_list=eps)
        train_perfs['fold']=fold
        train_perfs['eval_mode']=eval_mode
        train_perfs['subset']='train'
        train_perfs['postprocessing']='norm_prr'
        all_perfs.append(train_perfs) 
        del train_perfs

        th_list = optimize_thres(Y_train, P_train)
        hat_P_test = (P_test - np.array(th_list))/(1 - np.array(th_list))
        hat_P_test = hat_P_test.clip(lower=0,upper=1)
        filter_prr = (hat_P_test.rank(axis=1,method='min',ascending=False)<=Y_test.sum(axis=1).values.reshape(-1,1))*1+eps
        Y_hat_prr = hat_P_test*filter_prr   

        ##### MASK
        test_perfs = eval_species_perfs(Ys_hat=filter_prr,Ys_true=Y_test, th_list=eps)
        test_perfs['fold']=fold
        test_perfs['eval_mode']=eval_mode
        test_perfs['subset']='test'
        test_perfs['postprocessing']='norm_prr_mask'
        all_perfs.append(test_perfs)
        del test_perfs

        ##### CLIPPED PROBA
        test_perfs = eval_species_perfs(Ys_hat=Y_hat_prr,Ys_true=Y_test, th_list=eps)
        test_perfs['fold']=fold
        test_perfs['eval_mode']=eval_mode
        test_perfs['subset']='test'
        test_perfs['postprocessing']='norm_prr'
        all_perfs.append(test_perfs)  
       
```

### Training

<b> Losses </b>
- Species: binary_crossentropy, focal loss, weighted binary_crossentropy
- CWM, FDIS: mse, mae, mape, huber

```python
loss_dict = {
    'bce': lambda: tfk.losses.BinaryCrossentropy(from_logits=False),
    'fl': lambda: tfk.losses.BinaryFocalCrossentropy(gamma=1.0,from_logits=False),
    'wbce': lambda: tfk.losses.BinaryFocalCrossentropy(apply_class_balancing=True,gamma=0.0,from_logits=False),    
    'wfl': lambda: tfk.losses.BinaryFocalCrossentropy(apply_class_balancing=True,gamma=1.0,from_logits=False),
    'mse': lambda: tfk.losses.MeanSquaredError(),
    'mae': lambda: tfk.losses.MeanAbsoluteError(),
    'msle': lambda: tfk.losses.MeanSquaredLogarithmicError(),
    'mape': lambda: tfk.losses.MeanAbsolutePercentageError(),
    'huber': lambda: tfk.losses.Huber(delta=1.0)
}

archi_dict = {
    'glm' : [],
    'mlp1' : [512],
    'mlp2' : [512, 256],
    'mlp3' : [512, 256, 128]
}
```

```python
fold=1
eval_mode='interpolation'
_, (in_test, out_test), scalers = generate_dataset(metadata, Y, T, fold=fold,eval_mode=eval_mode,in_comm='mem',out_comm='obs', trait_scaler=None)

(in_train, out_train), _, scalers = generate_dataset(metadata, Y, T, fold=fold,eval_mode=eval_mode,in_comm='obs',out_comm='obs', trait_scaler=None)
```

```python
out_train['cwm']=out_train['cwm'].astype(np.float32)
out_train['fdis']=out_train['fdis'].astype(np.float32)
in_train[0][0]=in_train[0][0].astype(np.float32)
in_train[0][1]=in_train[0][1].astype(np.float32)
```

```python
species_loss = 'bce'
func_loss = 'msle'
archi_name = 'mlp1'
apply_prr = False
eps = 0.5
```

```python
out_dir = 'results/richesse_train_obsunw_test_degrad_obs_wlspecies10_lossweightedmomentsx10/%s_%s_%s/%s_%d'%(archi_name,species_loss,func_loss,eval_mode,fold)
os.makedirs(out_dir,exist_ok=True)
```

```python
set_seed(seed=1234)
```

```python
x=np.log(out_train['cwm'])
weigth_cwm = 10/(np.quantile(x, 0.95)-np.quantile(x,0.05))
x=np.log(out_train['fdis'])
weigth_fdis = 10/(np.quantile(x, 0.95)-np.quantile(x,0.05))
x=np.log(out_train['richness'])
weigth_richness = np.float32(10/(np.quantile(x, 0.95)-np.quantile(x,0.05)))
#x=out_train['species']
#weigth_species = 10/(np.quantile(x, 0.95)-np.quantile(x,0.05))
```

```python
trait_matrix = T.values
```

```python
archi = archi_dict.get(archi_name)
plant_model = MTNN(num_species=nspecies, num_traits=ntrait, trait_matrix = trait_matrix, apply_prr=apply_prr, optim_rich=True, hidden_units=archi, hidden_activation='relu', out_activation='sigmoid',dropout_rate=0.0,l1=0.0,l2=0.0,name='mtnn')
plant_model.compile(optimizer='adam',
              loss={'species':loss_dict.get(species_loss)(), 
                    'cwm': loss_dict.get(func_loss)(), 
                    'fdis': loss_dict.get(func_loss)(),
                    'richness' : loss_dict.get(func_loss)()
                   },
              loss_weights={'species':10,'cwm':weigth_cwm,'fdis':weigth_fdis, 'richness':weigth_richness},
              
              metrics={'species':[tfk.metrics.AUC(name='auc',multi_label=True,from_logits=False),
                                  tfk.metrics.AUC(name='aupr',curve='PR',multi_label=True,from_logits=False),
                                  tfk.metrics.Recall(thresholds=eps,name='recall'),
                                  tfk.metrics.Precision(thresholds=eps,name='precision')#,
                                  #tfk.metrics.BinaryAccuracy(threshold=eps,name='binary_accuracy')
                                 ],
                       
                       #'cwm':tfk.metrics.R2Score(class_aggregation="variance_weighted_average"),
                       #'fdis':tfk.metrics.R2Score(class_aggregation="variance_weighted_average"),
                       #'richness':tfk.metrics.R2Score()
                      })
```

```python
es_cbk= tfk.callbacks.EarlyStopping(monitor='val_loss',mode='min',patience=10,min_delta=0.0001, start_from_epoch=10)
plant_model.fit(x=in_train,y=out_train,batch_size=64,epochs=500,shuffle=True, 
                validation_data=(in_test, out_test), 
                callbacks=[es_cbk])
```

```python
prediction, species_perfs, comm_perfs, jaccard_perfs = evaluate_prediction(plant_model,
                                T_scaled = T,
                                data_train=(in_train, out_train), 
                                data_test=(in_test, out_test), 
                                splist=splist, tlist=tlist, 
                                method='raw', do_residual=False, eps=1e-6)

all_comm_perfs = [comm_perfs]
all_species_perfs = []
all_jaccard_perfs = []

for method in ['th05','opth','opth_norm','prr','norm_prr']:
    print(method)
    prediction_method, species_perfs, comm_perfs, jaccard_perfs = evaluate_prediction(plant_model,
                                    T_scaled = T,
                                    data_train=(in_train, out_train), 
                                    data_test=(in_test, out_test), 
                                    splist=splist, tlist=tlist, 
                                    method=method, do_residual=False, eps=1e-6)
    prediction_df = pd.DataFrame(prediction_method)
    prediction_df.to_csv('%s/prediction_%s.csv' %(out_dir, method))
    if comm_perfs is not None:
        comm_perfs['postprocess']=method
        all_comm_perfs.append(comm_perfs)
        
    if species_perfs is not None:     
        species_perfs['postprocess']=method
        all_species_perfs.append(species_perfs)
    if jaccard_perfs is not None : 
        jaccard_perfs = pd.DataFrame({'jaccard': jaccard_perfs[0]})
        jaccard_perfs['fold']=fold
        jaccard_perfs['eval_mode']=eval_mode
        jaccard_perfs['subset']='train'
        jaccard_perfs['postprocess']=method
        jaccard_perfs['prr'] = 'False'
        jaccard_perfs['archi']=archi_name
        jaccard_perfs['species_loss'] = species_loss
        jaccard_perfs['func_loss']=func_loss 
        jaccard_perfs = jaccard_perfs[['jaccard','postprocess','fold','eval_mode','prr','subset','archi','species_loss','func_loss']]
        all_jaccard_perfs.append(jaccard_perfs)

    
species_perfs_df = pd.concat(all_species_perfs, axis=0)
comm_perfs_df = pd.concat(all_comm_perfs, axis=0)
comm_perfs_df[['trait','indice']]=[c.split('_') if any([tr in c for tr in tlist]) else ['','richness'] if c=='richness' else ['',''] for c in comm_perfs_df.index]
comm_perfs_df = comm_perfs_df.reset_index()


species_perfs_df['fold']=fold
species_perfs_df['eval_mode']=eval_mode
species_perfs_df['prr']=apply_prr
species_perfs_df['archi']=archi_name
species_perfs_df['species_loss']=species_loss
species_perfs_df['func_loss']=func_loss

comm_perfs_df['fold']=fold
comm_perfs_df['eval_mode']=eval_mode
comm_perfs_df['prr']=apply_prr
comm_perfs_df['archi']=archi_name
comm_perfs_df['species_loss']=species_loss
comm_perfs_df['func_loss']=func_loss

jaccard_perfs_df = pd.concat(all_jaccard_perfs, axis=0)
```
